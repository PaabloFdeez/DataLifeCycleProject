```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("breastcancer.csv",sep=",")
```

```python
x_column = 'area_worst'
y_column = 'radius_mean'

# Crear el gr치fico de dispersi칩n
plt.scatter(df[x_column], df[y_column], label=f'{x_column} vs {y_column}')
plt.title(f'Gr치fico de dispersi칩n: {x_column} vs {y_column}')
plt.xlabel(x_column)
plt.ylabel(y_column)
plt.legend()
plt.show()
```
